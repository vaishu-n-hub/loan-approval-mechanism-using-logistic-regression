## Data Cleaning

### Checking the Missing Values

```python
df.isnull().sum()
```

#### First we will fill the Missing Values in "LoanAmount" & "Credit_History" by the 'Mean' & 'Median' of the respective variables.

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
```

### Let's confirm if there are any missing values in 'LoanAmount' & 'Credit_History'

```python
df.isnull().sum()
```

### Now, Let's drop all the missing values remaining.

```python
# Fill missing values for numeric columns
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())

# Fill missing values for categorical columns
df['Gender'] = df['Gender'].fillna('Unknown')
df['Married'] = df['Married'].fillna('No')
df['Dependents'] = df['Dependents'].fillna('0')  # Or use another strategy
df['Self_Employed'] = df['Self_Employed'].fillna('No')
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])

# Fill missing values in the 'Photo' column (for example, with a default photo path)
df['Photo'] = df['Photo'].fillna('C:\\Users\\Vaishnavi\\Desktop\\Loan-prediction-using-logistic-regression\\images\\default_photo.jpg')

# Now check the shape again
print("Shape after filling missing values:", df.shape)
```

### Let's check the Missing values for the final time!

```python
df.isnull().sum()
```

Here, we have dropped all the missing values to avoid disturbances in the model. The Loan Prediction requires all the details to work efficiently and thus the missing values are dropped.

### Importing Packages for Classification algorithms

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics 
import pandas 
```

### Splitting the data into Train and Test set

```python
X = df.iloc[1:542,1:12].values
y = df.iloc[1:542,12].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0)
```

### Logistic Regression (LR)

Logistic regression is a supervised learning classification algorithm used to predict the probability of a target variable. 

Mathematically, a logistic regression model predicts P(Y=1) as a function of X. It is one of the simplest ML algorithms that can be used for various classification problems such as spam detection, Diabetes prediction, cancer detection etc.

#### Sigmoid Function

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

X_imputed = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.3, random_state=0)

model = LogisticRegression()
model.fit(X_train, y_train)

lr_prediction = model.predict(X_test)
print('Logistic Regression accuracy = ', metrics.accuracy_score(lr_prediction, y_test))
```

```python
print("y_predicted",lr_prediction)
print("y_test",y_test)
```

```python
import cv2
import pandas as pd

df = pd.read_csv('C:\\Users\\Vaishnavi\\Desktop\\Loan-prediction-using-logistic-regression\\train.csv')

def capture_photo_for_loan(loan_id, df):
    cap = cv2.VideoCapture(0)  # Open the webcam

    # Check if the Loan_ID exists in the dataset
    if loan_id in df['Loan_ID'].values:
        print(f"Loan_ID {loan_id} found. Preparing to capture photo...")
        
        # Read a frame from the camera
        ret, frame = cap.read()

        if ret:
            # Create a unique filename based on Loan_ID
            image_path = f"C:\\Users\\Vaishnavi\\Desktop\\Loan-prediction-using-logistic-regression\\images\\user_photo_{loan_id}.jpg"
            
            # Save the captured photo
            cv2.imwrite(image_path, frame)
            print(f"Photo saved for Loan_ID {loan_id}: {image_path}")
            
            # Add the image path to the dataset
            df.loc[df['Loan_ID'] == loan_id, 'Photo'] = image_path
            
            # Show the captured photo for 1 second (1000 ms)
            cv2.imshow("Captured Photo", frame)
            cv2.waitKey(1000)  # Wait for 1 second before closing the window

        else:
            print(f"Error: Couldn't capture a photo for Loan_ID {loan_id}")

    else:
        print(f"Error: Loan_ID {loan_id} not found in the dataset")

    # Release the camera and close the window
    cap.release()
    cv2.destroyAllWindows()

    return df

# Prompt the user to enter a Loan_ID
loan_id_input = input("Enter the Loan_ID for which you want to capture a photo: ")

df = capture_photo_for_loan(loan_id_input, df)

# Save the updated dataset back to the same file (overwrite the original dataset)
df.to_csv('C:\\Users\\Vaishnavi\\Desktop\\Loan-prediction-using-logistic-regression\\train.csv', index=False)
print("Existing dataset updated with photo path and saved.")
```
